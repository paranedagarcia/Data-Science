# üìä Storytelling


## Introducci√≥n

En el √°mbito de la ciencia de datos, no basta con generar modelos precisos o visualizaciones impactantes. El verdadero valor surge cuando los resultados se convierten en historias comprensibles y accionables para distintos p√∫blicos: ejecutivos, clientes, comunidades o investigadores.

El **storytelling** es el arte de comunicar con datos de manera clara, convincente y memorable.

## Conceptos Generales del Storytelling en Datos

- Narrativa: La estructura que conecta los datos con un mensaje central.
- Contexto: Sin marco interpretativo, los datos pueden ser ambiguos.
- Visualizaci√≥n: Herramienta para dar forma y claridad a la historia.
- Emoci√≥n: El impacto emocional facilita la recordaci√≥n y acci√≥n.

üí° Ejemplo: Un gr√°fico que muestra un descenso en ventas es informaci√≥n. Pero si lo conectamos con una narrativa:
"Las ventas disminuyeron un 30% tras la competencia lanzar un nuevo producto" ‚Üí se convierte en historia.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Crear un dataset de ejemplo (ventas mensuales)
np.random.seed(42)
meses = pd.date_range("2024-01-01", periods=12, freq="M")
ventas = np.random.randint(80, 200, size=12)
df = pd.DataFrame({"Mes": meses, "Ventas": ventas})

# Plot de ventas mensuales
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Mes", y="Ventas", marker="o")
plt.title("üìä Ventas Mensuales - Storytelling con Datos", fontsize=14)
plt.xlabel("Mes")
plt.ylabel("Ventas (miles)")
plt.xticks(rotation=45)
plt.tight_layout()
#plt.savefig("/mnt/data/ventas_storytelling.png")
plt.show()

# Crear gr√°fico de barras para motivos de abandono (ejemplo ficticio)
motivos = ["Precio elevado", "Competencia", "Atenci√≥n deficiente", "Otros"]
porcentajes = [40, 25, 20, 15]
df2 = pd.DataFrame({"Motivo": motivos, "Porcentaje": porcentajes})

plt.figure(figsize=(8,6))
sns.barplot(data=df2, x="Motivo", y="Porcentaje", palette="viridis")
plt.title("üìâ Motivos de Abandono de Clientes", fontsize=14)
plt.ylabel("Porcentaje (%)")
plt.xticks(rotation=20)
plt.tight_layout()
#plt.savefig("/mnt/data/motivos_abandono.png")
plt.show()

# Crear heatmap de correlaciones (ejemplo ficticio en salud)
data = pd.DataFrame({
    "Edad": np.random.randint(20, 80, 100),
    "IMC": np.random.normal(25, 5, 100),
    "Colesterol": np.random.randint(150, 300, 100),
    "Riesgo": np.random.randint(0, 2, 100)
})

plt.figure(figsize=(8,6))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("üî• Matriz de Correlaci√≥n - Datos de Salud", fontsize=14)
plt.tight_layout()
#plt.savefig("/mnt/data/correlacion_salud.png")
plt.show()

```
