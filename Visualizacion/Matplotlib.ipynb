## Gr치fico simple con personalizaci칩n   

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# crear una serie de datos ficticios
# una lista de a침os desde 2011 a 2020
years = ["2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]

nivel1 = np.random.rand(10) * 100
nivel2 = np.random.rand(10) * 200 +100
nivel3 = np.random.rand(10) * 300 +200
nivel4 = np.random.rand(10) * 400 +300
nivel5 = np.random.rand(10) * 500 +400

# crear una figura y un eje
plt.plot(years, nivel1, label='Nivel 1')
plt.plot(years, nivel2, label='Nivel 2')
plt.plot(years, nivel3, label='Nivel 3')
plt.plot(years, nivel4, label='Nivel 4')
plt.plot(years, nivel5, label='Nivel 5')

# a침adir una leyenda
plt.legend()

# establecer el tipo de letra para todo el grafico
plt.rcParams['font.family'] = 'Times New Roman'

# a침adir etiquetas y t칤tulo
plt.xlabel('A침os de examen')
plt.ylabel('Puntajes')
plt.title('Puntajes a lo largo de los a침os', fontsize=16)

# personalizar eje de puntaje de 0 a 500 en incrementos de 50
plt.yticks(np.arange(0, 901, 50))

# activar cuadricula
plt.grid()

# activar marcas menores solo en el eje y
plt.minorticks_on()
plt.grid(which='minor', axis='y', linestyle=':', linewidth=0.5)

#establecer un tama침o de figura en pulgadas
plt.gcf().set_size_inches(12, 7)

# no mostrar ticks menores, solo mayores en el eje x en los a침os
plt.tick_params(axis='x', which='minor', bottom=False, top=False)

# mostrar el gr치fico
plt.show()
````

### Marcadores
El uso de marcadores permite diferenciar mayormente entre las series de un gr치fico.
[Gu칤a de marcas](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers)

````python
# establecer colores y marcas para cada l칤nea, esto sobreescribe los colores por defecto.
plt.plot(years, nivel1, label='Nivel 1', color='blue', marker='o')
plt.plot(years, nivel2, label='Nivel 2', color='orange', marker='s')
plt.plot(years, nivel3, label='Nivel 3', color='green', marker='^')
plt.plot(years, nivel4, label='Nivel 4', color='red', marker='+')
plt.plot(years, nivel5, label='Nivel 5', color='purple', marker='x')

# a침adir una leyenda
plt.legend()

# a침adir etiquetas y t칤tulo
plt.xlabel('A침os de examen')
plt.ylabel('Puntajes')
plt.title('Puntajes a lo largo de los a침os')

# personalizar eje de puntaje de 0 a 500 en incrementos de 50
plt.yticks(np.arange(0, 901, 50))

# activar cuadricula
plt.grid()

# activar marcas menores solo en el eje y
plt.minorticks_on()
plt.grid(which='minor', axis='y', linestyle=':', linewidth=0.5)

# establecer un tama침o de figura en pulgadas
plt.gcf().set_size_inches(12, 7)

# no mostrar ticks menores, solo mayores en el eje x en los a침os
plt.tick_params(axis='x', which='minor', bottom=False, top=False)

# mostrar el gr치fico
plt.show()
````

````python
# definir estilo de linea para cada serie de datos
plt.plot(years, nivel1, label='Nivel 1', color='blue', marker='o', linestyle='-')
plt.plot(years, nivel2, label='Nivel 2', color='orange', marker='s', linestyle='--')
plt.plot(years, nivel3, label='Nivel 3', color='green', marker='^', linestyle='-.')
plt.plot(years, nivel4, label='Nivel 4', color='red', marker='+', linestyle=':')
plt.plot(years, nivel5, label='Nivel 5', color='purple', marker='x', linestyle='-')

# a침adir una leyenda    
plt.legend()

# establecer un tama침o de figura en pulgadas
plt.gcf().set_size_inches(12, 7)

# mostrar el gr치fico
plt.show()
````

## Gr치fico de barras apiladas
En este tipo de gr치fico se define una variable "bottom" para establecer la apilaci칩n y orden de las series.

````python
plt.bar(years, nivel1, label='Nivel 1', color='blue')
plt.bar(years, nivel2, label='Nivel 2', color='orange', bottom=nivel1)
plt.bar(years, nivel3, label='Nivel 3', color='green', bottom=nivel1+nivel2)
plt.bar(years, nivel4, label='Nivel 4', color='red', bottom=nivel1+nivel2+nivel3)
plt.bar(years, nivel5, label='Nivel 5', color='purple', bottom=nivel1+nivel2+nivel3+nivel4)

# a침adir etiquetas y t칤tulo
plt.xlabel('A침os de examen')
plt.ylabel('Puntajes')
plt.title('Puntajes a lo largo de los a침os', fontsize=18)
plt.xlabel('A침os de examen', fontsize=14)
plt.ylabel('Puntajes', fontsize=14)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(loc='upper center', ncol=5)
plt.gcf().set_size_inches(12, 8)
# mostrar el gr치fico
plt.show()
````

## Torta o Pie

````python
plt.rcParams['font.family'] = 'Arial'

plt.pie(nivel1, labels=years, autopct='%1.1f%%', startangle=140)
plt.title('Distribuci칩n del Nivel 1 a lo largo de los a침os', fontsize=16)
plt.gcf().set_size_inches(6,6)
plt.show()
````

Crear un gr치fico que reune todos los niveles de puntaje en 2 columnas.

````python
fig, axs = plt.subplots(3, 2, figsize=(12, 18))
niveles = [nivel1, nivel2, nivel3, nivel4, nivel5]
titulos = ['Nivel 1', 'Nivel 2', 'Nivel 3', 'Nivel 4', 'Nivel 5']

for i, (nivel, titulo) in enumerate(zip(niveles, titulos)):
    row, col = divmod(i, 2)
    axs[row, col].pie(nivel, labels=years, autopct='%1.1f%%', startangle=140)
    axs[row, col].set_title(f'Distribuci칩n del {titulo}', fontsize=16)
    axs[row, col].figure.set_size_inches(12, 12)

# Eliminar el subplot vac칤o (칰ltima celda)
axs[2, 1].axis('off')

plt.tight_layout()
plt.show()
````

## Series

Definiremos datos aleatorios de temperatura a lo largo de un a침o. Estar치n con una distribuci칩n normal con una media de 25 grados.

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(120)  # Para reproducibilidad    
temperatura = np.random.normal(20, 5, 365)  
dias = np.arange(1, 366)
plt.figure(figsize=(12, 6))
plt.plot(dias, temperatura, linestyle='-', marker='o', markersize=3)
plt.title('Temperatura Diaria a lo Largo de un A침o', fontsize=16)
plt.xlabel('D칤a del A침o', fontsize=14)
plt.ylabel('Temperatura (춿C)', fontsize=14)
plt.xticks(np.arange(0, 366, 30), fontsize=10)
plt.yticks(np.arange(0, 41, 5), fontsize=10)
plt.grid()
plt.show()
````

- grafica con temperatura diaria
- distribucion de temperaturas

````python
fig, ax = plt.subplots(1, 2)

ax[0].plot(temperatura, label="Temperatura diaria")
ax[0].set_xlabel("D칤a")
ax[0].set_ylabel("Temperatura (춿C)")
ax[0].set_title("Temperatura diaria")
ax[0].legend()

ax[1].hist(temperatura, bins=10, edgecolor="black")
ax[1].set_xlabel("Temperatura (춿C)")
ax[1].set_ylabel("Frecuencia")
ax[1].set_title("Distribuci칩n")

#plt.tight_layout()
plt.gcf().set_size_inches(12, 6)
plt.show()
````

````python
# grafico de temperaturas maximas y minimas
maximas = pd.read_csv('../data/maximas.csv', parse_dates=['Fecha'], dayfirst=True)
minimas = pd.read_csv('../data/minimas.csv', parse_dates=['Fecha'], dayfirst=True)

plt.figure(figsize=(12, 6))
plt.plot(maximas['Fecha'], maximas['Temperatura'], label='Temperaturas M치ximas', color='red')
plt.plot(minimas['Fecha'], minimas['Temperatura'], label='Temperaturas M칤nimas', color='blue')
plt.xlabel('Fecha', fontsize=14)
plt.ylabel('Temperatura (춿C)', fontsize=14)
plt.title('Temperaturas M치ximas y M칤nimas Diarias en 2024', fontsize=16)
plt.xticks(rotation=90, fontsize=10)

# personalizar eje de puntaje de -5 a 30 en incrementos de 5 grados
plt.yticks(np.arange(-5, 31, 5), fontsize=10)

plt.xticks(maximas['Fecha'][::31], rotation=90, fontsize=9)

# activar cuadricula
plt.grid()
# activar marcas menores solo en el eje y
plt.minorticks_on()
plt.grid(which='minor', axis='y', linestyle=':', linewidth=0.5)

plt.legend()
plt.grid()
plt.show()
````

````python
# Crear un nuevo DataFrame con columnas 'Fecha' y 'Temperatura'
fechas = []
temperaturas = []

meses = maximas.columns[1:]  # Ignora la columna 'Dia'

for mes in meses:
    for idx, row in maximas.iterrows():
        dia = row['Dia']
        temp = row[mes]
        # Solo agregar si el valor de temperatura no es nulo y es num칠rico
        if pd.notnull(temp) and isinstance(temp, (int, float, np.float64)):
            fechas.append(f"{dia:02d}-{mes}-2024")
            temperaturas.append(temp)

df_temperaturas = pd.DataFrame({'Fecha': fechas, 'Temperatura': temperaturas})
df_temperaturas.head()
````
